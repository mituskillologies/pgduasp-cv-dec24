##### Basic Histogram

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
marks = np.array([56,66,56,87,67,62,28,65,48,89,82,66,59,52,60,63,46,67,88])
```

```python
plt.title('Histogram of Marks')
plt.xlabel('Range of Marks')
plt.xticks(range(0,101,10))
plt.grid()
plt.ylabel('Count')
plt.hist(marks, bins = range(0,101,10), color = 'gray');
```

```python
import seaborn as sns
```

```python
sns.histplot(marks, kde=True);
```

```python
img = cv2.imread('images/lenna.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

##### Color Image Histogram

```python
img = cv2.imread('images/lenna.png')
```

```python
b, g, r = cv2.split(img)
```

```python
plt.figure(figsize=(16,9))
plt.title("Color Image Histogram")
plt.subplot(1,3,1)
plt.xticks(range(0,256,20))
plt.hist(b.flatten(), bins=range(0,256), color = 'b');
plt.subplot(1,3,2)
plt.xticks(range(0,256,20))
plt.hist(g.flatten(), bins=range(0,256), color = 'g');
plt.subplot(1,3,3)
plt.xticks(range(0,256,20))
plt.hist(r.flatten(), bins=range(0,256), color = 'r');
```

##### Stretching and Shrinking

```python
# Load the image
image = cv2.imread("images/lenna.png")

# Define new dimensions (larger than original)
new_width, new_height = 512, 200

# Stretch the image using interpolation
stretched_image = cv2.resize(image, dsize=(new_width, new_height), 
                             interpolation=cv2.INTER_CUBIC)

cv2.imshow("Stretched Image", stretched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load the image
image = cv2.imread("images/lenna.png")

# Define new dimensions (larger than original)
new_width, new_height = 200, 512

# Stretch the image using interpolation
stretched_image = cv2.resize(image, dsize=(new_width, new_height), 
                             interpolation=cv2.INTER_CUBIC)

cv2.imshow("Stretched Image", stretched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Histogram Equalization

```python
# import Numpy 
import numpy as np 
  
# read a image using imread 
img = cv2.imread('images/lenna.png') 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
# creating a Histograms Equalization of a image using cv2.equalizeHist() 
equ = cv2.equalizeHist(gray) 
  
# show image input vs output 
cv2.imshow('Equalized image', equ) 
  
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
