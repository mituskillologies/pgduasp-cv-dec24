##### Averaging Kernal

```python
import cv2
import numpy as np

# Load an image
image = cv2.imread("images/lenna.png")

# Define a 3x3 averaging kernel
kernel = np.ones((5, 5), np.float32) / 25

# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)

# Show the original and smoothed images
cv2.imshow("Original", image)
cv2.imshow("Smoothed", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel
```

```python
final_image = np.vstack([image, smoothed_image])
```

```python
cv2.imshow("Final", final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load an image
image = cv2.imread("images/sun.png")
kernel = np.ones((7, 7), np.float32) / 49
smoothed_image = cv2.filter2D(image, -1, kernel)
final_image = np.hstack([image, smoothed_image])
cv2.imshow("Smoothed", final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Gaussian Kernel

```python
kernel_size = 7  # Adjust as needed
sigma = 0.05 * ((kernel_size - 1) * 0.5 - 1) + 2.35  # Gaussian standard deviation
kernel = cv2.getGaussianKernel(kernel_size, sigma)
```

```python
kernel
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(7), kernel)
```

```python
import seaborn as sns
sns.kdeplot(kernel);
```

```python
# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)

# Show the original and smoothed images
cv2.imshow("Original", image)
cv2.imshow("Smoothed", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
final_image = np.vstack([image, smoothed_image])

cv2.imshow("Final", final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# averaging kernel
avg_kernel = np.ones((5, 5), np.float32) / 25

# gaussian kernel
sigma = 0.05 * ((7 - 1) * 0.5 - 1) + 2.35  # Gaussian standard deviation
gauss_kernel = cv2.getGaussianKernel(7, sigma)
```

```python
avg_image = cv2.filter2D(image, -1, avg_kernel)
gauss_image = cv2.filter2D(image, -1, gauss_kernel)
```

```python
comp_image = np.hstack([image, avg_image, gauss_image])
```

```python
cv2.imshow("Final", comp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Median Filter

```python
img = cv2.imread('images/lenna.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernel_size = 15
median = cv2.medianBlur(gray, kernel_size)
```

```python
median.shape
```

```python
plt.figure(figsize=(12,9))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(gray, cmap = 'gray')
plt.subplot(1,2,2)
plt.title('Median Blurred Image')
plt.imshow(median, cmap = 'gray')
```

```python
median3 = cv2.medianBlur(gray, 3)
median5 = cv2.medianBlur(gray, 5)
median7 = cv2.medianBlur(gray, 7)
median9 = cv2.medianBlur(gray, 9)
median11 = cv2.medianBlur(gray, 11)
median13 = cv2.medianBlur(gray, 13)
median15 = cv2.medianBlur(gray, 15)
```

```python
plt.figure(figsize=(15,7))
plt.subplot(2,4,1)
plt.title('Original')
plt.imshow(gray, cmap = 'gray')
plt.subplot(2,4,2)
plt.title('3-Median')
plt.imshow(median3, cmap = 'gray')
plt.subplot(2,4,3)
plt.title('5-Median')
plt.imshow(median5, cmap = 'gray')
plt.subplot(2,4,4)
plt.title('7-Median')
plt.imshow(median7, cmap = 'gray')
plt.subplot(2,4,5)
plt.title('9-Median')
plt.imshow(median9, cmap = 'gray')
plt.subplot(2,4,6)
plt.title('11-Median')
plt.imshow(median11, cmap = 'gray')
plt.subplot(2,4,7)
plt.title('13-Median')
plt.imshow(median13, cmap = 'gray')
plt.subplot(2,4,8)
plt.title('15-Median')
plt.imshow(median15, cmap = 'gray')
```

##### Bilateral Filter

```python
bilateral = cv2.bilateralFilter(gray, d=9, sigmaColor=95, sigmaSpace=75)  
# Adjust parameters

plt.figure(figsize=(10, 6))
plt.subplot(121), 
plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), 
plt.imshow(bilateral, cmap='gray'), plt.title('Bilateral Filtered Image')
plt.show()
```
